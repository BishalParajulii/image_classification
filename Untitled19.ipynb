```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import datasets , layers , models
```

```python
#to load data directly from keras  use this
(training_image ,training_labels ) , (testing_image , testing_labels) = datasets.cifar10.load_data()
```

```python
training_image , testing_image = training_image / 255 , testing_image / 255
```

```python
class_names = ['plane' , 'car' , 'bird' , 'car' , 'deer' , 'dog' , 'frog' , 'horse' , 'ship' , 'truck']
```

```python
for i in range(16):
  plt.subplot(4,4,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(training_image[i] , cmap=plt.cm.binary)
  plt.xlabel(class_names[training_labels[i][0]])

plt.show()
```

```python
training_image = training_image[:20000]
training_labels = training_labels[:20000]
testing_image = testing_image[:20000]
testing_labels = testing_labels[:20000]
```

```python
#building Neural Network
model = models.Sequential()

#defining input layers
model.add(layers.Conv2D(32 , (3,3) , activation = 'relu' , input_shape = (32 , 32 , 3)))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(64 , (3,3) , activation = 'relu'))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(64 , (3,3) , activation = 'relu'))
model.add(layers.Flatten())

model.add(layers.Dense(64 , activation = 'relu'))
model.add(layers.Dense(64 , activation = 'softmax'))

model.compile(optimizer = 'adam' , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])

model.fit(training_image , training_labels , epochs = 20 , validation_data = (testing_image , testing_labels))
```

```python
model.save("image_classifier.model")
```
